```python
from typing import Sequence
from kipy import KiCad
from kipy.board import Board
from kipy.board_types import FootprintInstance

kicad = KiCad()
print(f"Connected to KiCad {kicad.get_version()}")

board: Board = kicad.get_board()

footprints: Sequence[FootprintInstance] = board.get_footprints()

for footprint in footprints:
    ref: str = footprint.reference_field.text.value
    # Check if ref is in the range D304-D344
    if ref.startswith("D") and ref[1:].isdigit():
        num = int(ref[1:])
        if 138 <= num <= 299:
            new_num: int = num + 1
            new_ref: str = f"D{new_num}"
            print(f"Changing {ref} -> {new_ref}")
            footprint.reference_field.text.value = new_ref
        else:
            print(ref)
    else:
        print(ref)

board.update_items(items=footprints)
```

```python
from kipy import KiCad
from kipy.board import Board
from kipy.util import to_mm

kicad = KiCad()
print(f"Connected to KiCad {kicad.get_version()}")

board: Board = kicad.get_board()

# --- Footprints with value "WS2812B" ---
all_footprints = board.get_footprints()
ws2812b_footprints = []

print("\nProcessing footprints to find 'WS2812B'...")
for footprint in all_footprints:
    if (footprint.value_field.text.value == "WS2812B"):
        ws2812b_footprints.append(footprint)
print(f"\nFound {len(ws2812b_footprints)} footprints with value 'WS2812B':")


# --- Text items with font "Gotham Narrow Medium" ---
all_text_items = board.get_text()
gotham_narrow_text_items = []

print("\nProcessing text items to find font 'Gotham Narrow Medium'...")
for text_item in all_text_items:
    if (hasattr(text_item, 'attributes') and
        hasattr(text_item.attributes, 'font_name') and
        text_item.attributes.font_name == "Gotham Narrow Medium" and
        text_item.layer == 40): # F.SilkS layer
        gotham_narrow_text_items.append(text_item)


print(f"\nFound {len(gotham_narrow_text_items)} text items with font 'Gotham Narrow Medium':")

for text_item in gotham_narrow_text_items:
    nearest_footprint = None
    min_distance = float('inf')
    for footprint in ws2812b_footprints:
        distance = to_mm(
            ((text_item.position.x - footprint.position.x) ** 2 +
             (text_item.position.y - footprint.position.y) ** 2) ** 0.5
        )
        if distance < min_distance:
            min_distance = distance
            nearest_footprint = footprint
    
    if nearest_footprint and round(min_distance,2) != 2.0:
        print(f"'{text_item.value}' is nearest to {nearest_footprint.reference_field.text.value}' @{min_distance:.2f}mm")
        
```
